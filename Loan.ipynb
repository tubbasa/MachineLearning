```python
loanDf.dropna(subset=["Credit Score"],inplace=True)
```

```python
loanDf['Loan Status'] = [1 if s=='Fully Paid' else 0 for s in loanDf['Loan Status']]
```

```python
loanDf['Credit Score'] = [int(x) for x in loanDf['Credit Score']]
loanDf['Annual Income'] = [int(x) for x in loanDf['Annual Income']]
loanDf['Years of Credit History'] = [int(x) for x in loanDf['Years of Credit History']]
```

```python
features =['Credit Score','Current Loan Amount','Annual Income','Home Ownership']
X = loanDf[features]
X = pd.get_dummies(X)
Y =loanDf['Loan Status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.20,random_state=4)  #train_size=0.8
```

```python
model = RandomForestClassifier(n_estimators=200, max_depth=3, random_state=2)
#estimator : tahmin etmek için ağaç sayısı
#max_depth : tahmin ağaçlarınızın ne kadar derin olmasını istediğimiz. Max depth küçük omalı
model.fit(X, Y)
prediction_for_rf = model.predict(x_test)
```

```python
accuracy_score(y_test, prediction_for_rf)
```

```python
confusion_matrix(y_test,prediction_for_rf)
```

```python
classification_report(y_test,prediction_for_rf)
```

```python
precision_score(y_test,prediction_for_rf, average='weighted')
```

```python
f1_score(y_test,prediction_for_rf, average='weighted')
```

```python
recall_score(y_test,prediction_for_rf, average='weighted')
```
