```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.isnull().sum()
```

```python
X = df[["Customers.id","Orders.total"]]
```

```python
df.columns
```

```python
df.sample(20)
```

```python
X_time = df[["Orders.payment_date","Orders.shipped_date"]]
X_time
```

```python
X_time["Orders.shipped_date"].fillna(X_time["Orders.payment_date"],inplace=True)
X_time["Orders.payment_date"].fillna(X_time["Orders.shipped_date"],inplace=True)
X_time.fillna(0,inplace=True)
X_time.isnull().sum()
```

```python
X_time_diff =abs(X_time["Orders.shipped_date"] - X_time["Orders.payment_date"]) / 1000
X_time_diff[X_time_diff < 0]
```

```python
X["shipped_time"] = X_time_diff
```

```python
df.isnull().values.all(axis=0)
```

```python
df.isnull().sum().value_counts
```

```python
X["Orders.total"].isnull().sum()
```

```python
X
```

```python
kmeans = KMeans(n_clusters=3) # You want cluster the passenger records into 2: Survived or Not survived
k_fit = kmeans.fit(X)
```

```python
merkezler = k_fit.cluster_centers_
merkezler
```

```python
X.iloc[:,0]
```

```python
kumeler = k_fit.labels_
kumeler
```

```python
plt.scatter(X.iloc[:,0],X.iloc[:,1],c=kumeler,s=50,cmap='viridis')

plt.scatter(merkezler[:,0],merkezler[:,1],c="red",s=200,alpha=0.5)
```

```python
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize']=(16,9)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X.iloc[:,0],X.iloc[:,1],X.iloc[:,2],c=kumeler)
ax.scatter(merkezler[:,0],merkezler[:,1],merkezler[:,2],marker='*',c='#050505',s=1000)
```
